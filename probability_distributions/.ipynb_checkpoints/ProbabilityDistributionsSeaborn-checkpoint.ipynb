# Probability Distributions
If I ask you think of a random number, there is literally an infinite amount of different values you might give me. But when applying the term "random" to solving real-world problems, our experience gives us an intuition that some values are more likely to occur than others. For instance, if you go to the pharmacy right now to pick up a prescription, we might not know the exact amount of time you'll end up having to wait in line and we could simulate this using a random value drawn from a **probability distribution**. These probability draws on our knowledge of past events to help us narrow down the range of an uncertain value. Thus, based on your previous expereinces at that pharmacy, you know that you typically wait in line eleven minutes, you've never gotten out in less than five minutes, and one particularly bad day you had to wait eighteen minutes. Under no circumstance would you expect to wait 300 minutes. Probability distributions use information about past events to gain  much better idea of what value the unknown parameter is likely to take.

- values of an unknown parameter vary based on the underlying probability distribution
- a probability dsitribution **p(x)** indicates the likelihood that the random variables takes a particular value of x
- describe the dispersion of the values that a random variable can take

## Two types
1. **Discrete probability distributions** for discrete variables (aka probability mass functions)
    - the sum of all probabilities must equal 1
    - can assume a discrete  number of values. Discrete numbers are finite "whole", for instance the number of patients a doctor sees per hour (she might see 11 or 12 patients, but nothing in between)
    - Types of discrete distributions:
        - Binary
        - Poisson
        - Categorical

2. **Probability density functions** for continuous variables (aka continuous probability functions)
    - the sum of the area under the curve equals 1
    - if the variable can assume an infinitie number of values between the max and min
    - with continuous data, a probability indicates the likelihood that a value will fall within a range
    - most continuous probability distributions have between 1-3 parameters that represent essential properties and ultimately define its shape
    - "distribution parameters are values that apply to entire populations. Unfortunately, its generally impossible to measure an entire population. So you can use random samples to calculate estimates of these parameters"
    
## Special Probability Distributions for Statistical Hypothesis Testing
- use sampling distributions and the distributions of test statistics to determine if the results are statistically signigicant
- we draw a large number of random samples from a population and calculate their mean, graph their distribution. This type of **sampling distribution** helps you determine the likelihood of getting different sample outcomes. 

## How to do determine which discrete probability distribution fits your data?
1. If data is binary, check the assumptions
- no goodness-of-fit test needed
2. For count data, use a Posisson goodness-of-fit test
- perform a Chi-square goodness-of-fit test to decide if it fits the distribution
- test compares the theoretical versus the observed frequencies of the values
- if the difference between the theoretical and the observed frequencies is significant, then the data does not follow that particular distribution
3. for categorical data, use the chi-square goodness-of-fit test

The location `loc` keyword specifies the mean. 
The scale `scale` keyword specifies the standard deviation.

```python
#Define a function to generate probability distribution
def plot_distribution(distribution, num=1000, mean=0, sd=1,**kwargs):
    
    """
    name: name of distribution
    num: number of samples to generate
    mean: loc
    sd: scale aka standard deviation
    kwargs:
        p: for specifying bernoulli and binomial (should be between 0-1)
        n: for specifying binom 
        mu: for specifying poisson
        a: for specifying gamma
    """
    p = kwargs.get('p', None)
    n = kwargs.get('n', None)
    mu = kwargs.get('mu', None)
    a = kwargs.get('a', None)
    
    dist_name = (type(distribution).__name__)
    if dist_name == 'uniform_gen':
        data = distribution.rvs(size=num, loc = mean, scale=sd)
        
    elif dist_name == 'norm_gen':
        data = distribution.rvs(size=num, loc = mean, scale=sd) 
        
    elif dist_name == 'bernoulli_gen':
        data = distribution.rvs(size=num, p=p)
        
    elif dist_name =='binom_gen':
        data = distribution.rvs(n=n,p=p,size=num)
        """ 
        Arguments for binomial distribution:
            p: probability of success
            n: number of independent experiments per trial
            size: number of trials
        """
    elif dist_name =='poisson_gen':
        data = distribution.rvs(mu, size=num)
        
    elif dist_name =='gamma_gen':
        data = distribution.rvs(a, size=num)
        
    else: 
        print("Sorry, I don\'t recognize that distribution.'")
        
        
    ax = sns.distplot(data,
                   kde=False,
                  color='teal')

    ax.set_xlabel(dist_name, fontsize=13)
    ax.set_ylabel('Frequency', fontsize=13)
    ax.set_title(label= 'Plot of ' + dist_name+ ' distribution \n (' + str(num) + ' samples)', fontsize=15)
```

```python
plot_distribution(norm, num=10000, mean=50, sd=200,)
```

```python
plot_distribution(uniform, num=10000, mean=30, sd=35,)
```

```python
plot_distribution(bernoulli, num=1500,p=0.4)
```

```python
plot_distribution(binom, num=1500, n=10, p=0.2)
```

```python
plot_distribution(gamma, a=1.5, num=1500)
```

```python
plot_distribution(poisson, num=1500, mu=6)
```
